## Normalizing Cleaned Dataset 

```python
import pandas as pd
import numpy as np
from numpy import random
raw_data = pd.read_csv('../rodpump_cleaned.csv')
raw_data = raw_data.drop('lifetime_end', axis=1)
```

## Splitting Up Data into the Failure Types

### Setting the  max size to the size of Rods, because there are the least number of Rod failures

```python
SRP = pd.DataFrame(raw_data[raw_data['FAILURETYPE'] == 'Sucker Rod Pump'])
Rods = pd.DataFrame(raw_data[raw_data['FAILURETYPE'] == 'Rods'])
Tubing = pd.DataFrame(raw_data[raw_data['FAILURETYPE'] == 'Tubing'])
normalized_data = Rods
maxSize = len(Rods.index)
```

```python
# random number gen to select row in data frame 1-len(Rods.index)
#put that row into new df
#remove row from current df
#shuffle df
#repeat for maxSize


def normalize(df,maxSize):
    global normalized_data
    for i in range(1,maxSize+1):
        randRowNum = random.randint(0,len(df.index))
        normalized_data = normalized_data.append(df.iloc[randRowNum,:])
        df = df.drop(df.index[randRowNum])

        #shuffle rows in df
        df = df.sample(frac=1).reset_index(drop=True)
    
    
```

```python
normalize(SRP, maxSize)
normalize(Tubing, maxSize)

#shuffle rows in new df
normalized_data.sample(frac=1).reset_index(drop=True)
```

```python
normalized_data.to_csv('../normalized_data.csv',index=False,header=True)
```
